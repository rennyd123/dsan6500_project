### Assigning Pictures to Class Categories

```python
import pandas as pd
from tqdm import tqdm
import os
from shutil import copy2

df = pd.read_csv("../Data_Entry_2017_v2020.csv")
df.head()

df = df[["Image Index", "Finding Labels"]]
df.columns = ["Image Index", "Finding Labels"]

image_base_folder = "../data/all_images"
output_folder = "../data/output_images"

os.makedirs(output_folder, exist_ok=True)

def copy_images_to_class_folders(row):
    image_name = row["Image Index"]
    classes = row["Finding Labels"].split("|")

    for sub_dir in range(1, 13):
        image_folder = os.path.join(image_base_folder, f'images_{sub_dir}', 'images')
        src = os.path.join(image_folder, image_name)
        if os.path.exists(src):
            break

    for cls in classes:
        class_folder = os.path.join(output_folder, cls.strip())
        os.makedirs(class_folder, exist_ok=True)

        dst = os.path.join(class_folder, image_name)

        copy2(src, dst)

for index, row in tqdm(df.iterrows(), total = df.shape[0], desc = "Processing images"):
    copy_images_to_class_folders(row)

print("Images have been successfully separated into class folders. ")
```
